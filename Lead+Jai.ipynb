# ***Building the Logistic Regression***

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

# Drop the 'Unnamed: 0' column
# data = data.drop(columns=['Unnamed: 0'])

# Define the features and target variable
X = df.drop(columns=['converted'])
y = df['converted']

# Normalize the feature data
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Initialize the Logistic Regression model
logreg = LogisticRegression()

# Train the model
logreg.fit(X_train, y_train)

# Predict on the test set
y_pred = logreg.predict(X_test)

# Evaluate the model
report = classification_report(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

# Display the classification report and confusion matrix
print('Classification Report:\n', report)
print('Confusion Matrix:\n', conf_matrix)
```

## ***The model shows a good balance between precision and recall, with an overall accuracy of 82%. The next step would be to assign lead scores based on the model's predicted probabilities.***

```python
# Calculate the predicted probabilities of conversion for each lead
y_scores = logreg.predict_proba(X_scaled)[:, 1]  # Get the probability of the positive class

# Convert probabilities to scores out of 100
lead_scores = y_scores * 100

# Add the lead scores to the original dataframe
df['lead_score'] = lead_scores

# Display the head of the dataframe with the lead scores
print(df[['lead number', 'lead_score']].head())
```

### This code calculates the predicted probabilities of conversion for each lead using the predict_proba method of a logistic regression model (logreg). It extracts the probability of the positive class (conversion) and converts these probabilities to scores out of 100. The lead scores are then added to the original DataFrame under the 'lead_score' column, and the head of the DataFrame with the lead scores is displayed.
